```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Load the CSVs
df = pd.read_csv ('03_complete-data-pop-days.csv', index_col='date', parse_dates=True)

print(df.head())
print(df.dtypes)
```

```python
#The same as plot
df.groupby(['date'])["Confirmed",'Deaths','Recovered'].sum().plot()
```

```python
df_germany=df[df['country'] == 'Germany']
#The same as plot
df_germany.groupby(['date'])["Confirmed",'Deaths','Recovered'].sum().plot()
```

```python
df_gb=df[df['country'] == 'United Kingdom']
#The same as plot
df_gb.groupby(['days'])["Confirmed",'Deaths','Recovered'].sum().plot()
```

```python
df_it=df[df['country'] == 'Italy']
#The same as plot
df_it.groupby(['days'])["Confirmed",'Deaths','Recovered'].sum().plot()
```

```python
#Filter
df_germany=df[df['country'] == 'Germany']

#print(df_germany)

dfnew=df_germany.groupby(['days'])["% Confirmed","% Active","% Recovered"].sum()

#print(dfnew)

# Add title and axis names

dfnew.plot()
plt.title('COVID 2019 - Germany')
plt.ylabel('% of Population')
plt.xlabel('Date')
plt.xticks(rotation=0)
```
