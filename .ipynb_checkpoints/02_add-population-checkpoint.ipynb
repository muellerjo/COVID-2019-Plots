```python
#After the Union of the CSVs, we will add the population of the countries to the completed dataset

#First we load the dataset with the numbers of world population

import pandas as pd
import numpy as np

#Load the CSVs
covid_cases = pd.read_csv ('python-complete-data.csv')
population = pd.read_csv ('worldpopulation/data.csv')

#Replace country names, because they er different in the population file
population["name"]= population["name"].replace('China', "Mainland China")
population["name"]= population["name"].replace('United States', "US")


#Merging both data
df = pd.merge(covid_cases, 
                  population[['name', 'pop2020']],
                  left_on='country',
                  right_on='name',
                  how='left')


df['date'] = df['date'].astype('datetime64')

def add_days(dataframe):
    print("add_days started")
    dates = dataframe.index.get_level_values(1)
    return dataframe.assign(days=dates - dates[0] + pd.Timedelta(days=1))

df = df.set_index(["country", "date"]).sort_index()
print(df)
result_df = df[df["Confirmed"] > 0].groupby(level=0).apply(add_days)
print(result_df)

#-----------------------------------------------------------------------------------
#Calculate KPIs
#-----------------------------------------------------------------------------------

#df['confirmed_percent'] = (df.Confirmed/(df.pop2020*1000))
#df['% Confirmed'] = df.confirmed_percent*100
#df['confirmed_per_Thousand'] = df.Confirmed/(df.pop2020)

#df['mortality_percent'] = df.Deaths / (df.Confirmed)
#df['% Mortality'] = df.mortality_percent*100

#df['% Death'] = (df.mortality_percent/(df.pop2020*1000))*100

#df['active'] = df.Confirmed - (np.nan_to_num(df.Recovered))
#df['active_percent'] = (df.Confirmed - (np.nan_to_num(df.Recovered)))/(df.pop2020*1000)
#df['% Active'] = df.active_percent*100
#df['active_per_Thousand'] = (df.Confirmed - (np.nan_to_num(df.Recovered)))/df.pop2020

#df['recovered_per_Thousand'] = df.Recovered/df.pop2020
#df['recovered_percent'] = df.Recovered/(df.pop2020*1000)
#df['% Recovered'] = df.recovered_percent*100

#-----------------------------------------------------------------------------------
#       Save CSV
#-----------------------------------------------------------------------------------


#Save to csv-file
print("\n Saving to CSV.............................................................\n")
df.to_csv('02_complete-data-pop.csv')

df.head()
```
